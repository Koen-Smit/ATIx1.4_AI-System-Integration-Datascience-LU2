WAT IS NORMAALVERDELING?


- Een normaalverdeling is een symmetrische verdeling van data rond het gemiddelde, waarbij de meeste waarden dicht bij het gemiddelde liggen en afnemen naarmate ze verder afwijken.

```python
import seaborn as sns

penguins = sns.load_dataset("penguins")
penguins.head() # first 5

penguins['sex'].unique()
penguins['sex'].value_counts(dropna=True).plot(kind='bar') # "dropna" drops NaN from data
```
