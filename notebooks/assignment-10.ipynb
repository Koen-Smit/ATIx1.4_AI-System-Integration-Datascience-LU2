#### Assumption:
Ik verwacht dat er weinig correlations zijn omdat het vrij random is wat een pokemon voor stats heeft.

```python
# Laad benodigdheden in voor de rest van dit bestand.
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
pokemon = pd.read_csv('Pokemon.csv', sep=',')
```

```python
numeric_pokemon = pokemon.select_dtypes(include='number')
numeric_pokemon = numeric_pokemon.drop(columns=['#', 'Generation'])

corr = numeric_pokemon.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', center=0, fmt='.2f', linewidths=0.5)
plt.title("Correlation of Pokemon Numerical")
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(18, 5))

sns.scatterplot(data=pokemon, x='Total', y='Sp. Atk', ax=axes[0])
axes[0].set_title('Total vs Sp. Atk stat points')
axes[0].set_xlabel('Total')
axes[0].set_ylabel('Sp. Atk')
axes[0].grid(True)

sns.scatterplot(data=pokemon, x='Total', y='Attack', ax=axes[1])
axes[1].set_title('Total vs Attack stat points')
axes[1].set_xlabel('Total')
axes[1].set_ylabel('Attack')
axes[1].grid(True)

plt.tight_layout()
plt.show()
```

#### Findings:
Beide grafieken tonen een vrij sterke positieve correlatie: hoe hoger de Total stat, hoe hoger meestal ook de Sp. Atk en Attack  opzich logisch, "Total" is een verzameling van alles opgeteld, en bij veel pokemon gaat het vooral om de attack-based stats.
