```python
# Laad benodigdheden in voor de rest van dit bestand.
import pandas as pd
pokemon = pd.read_csv('Pokemon.csv', sep=',')
```

### Column 1, Attack:

```python
pokemon['Attack'].plot(kind='hist', bins=10)
```

#### Findings:
De verdeling van de waarden in de kolom "Attack" is niet gelijk verdeeld(uniform) en ook niet netjes in het midden zoals bij een normale verdeling. Er zijn veel Pokémon met een lage tot gemiddelde aanval en maar een paar met een hele hoge aanval. Dit zorgt voor een lange "staart" naar rechts. Op de website van NIST heb ik gekeken en ik vind dat dit patroon het best past bij een lognormale verdeling.

### Column 2, Speed:

```python
pokemon['Speed'].plot(kind='hist', bins=10)
```

#### Findings:
Bij de kolom "Speed" zie je hetzelfde patroon: veel Pokémon hebben een lage tot gemiddelde snelheid, en maar weinig zijn echt snel. De verdeling loopt dus als het waren scheef naar rechts, met een piek aan de linkerkant. Dit lijkt ook op een lognormale verdeling, zoals uitgelegd op de website van NIST.
