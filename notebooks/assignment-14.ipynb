**Perform a bivariate analysis on at least 1 combination of 2 columns with categorical data in the dataset that you chose in portfolio assignment 4(pokemon.csv):**

- Do you expect their to be a correlation between the two columns?
- Create a contingency table. Do you observe different ratios between categories here?
- Create a bar plot for this contingency table. Do you observe different ratios between categories here?
- Do a chi-squared test. What does the result say? What's the chance of there being a correlation between the two columns?

---


```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency

pokemon = pd.read_csv('pokemon.csv')
pokemon[['Type 1', 'Legendary']].head()
```

#### Verwachtingen:
ik verwacht een verband: bepaalde types (zoals Dragon, Psychic of Steel) staan bekend om hoe "sterk" ze zijn als Pokémon, en zouden dus relatief meer Legendary Pokémon kunnen bevatten.

```python
ct_type_legendary = pd.crosstab(pokemon['Type 1'], pokemon['Legendary'])
print(ct_type_legendary)
```

```python
ct_type_legendary.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='Set2')
plt.title('Type 1 vs Legendary Status')
plt.ylabel('Aantal Pokémon')
plt.xlabel('Type 1')
plt.xticks(rotation=45)
plt.legend(title='Legendary')
plt.tight_layout()
plt.show()
```

#### Findings:
- Types zoals: psychic en dragon hebben relatief veel legendary pokemon vergeleken met andere types.
- Bug of normal types bevatten geen legendaries.
- De verhouding verschilt duidelijk per type.

```python
#Chi-Squared Test
chi2, p, dof, expected = chi2_contingency(ct_type_legendary)
print(f"Chi2: {chi2:.2f}, p-value: {p:.4f}")
```

#### Findings:
- de p-waarde ligt onder 0,05 en daarom is er dus een significant verband tussen type 1 en of een pokemon legendary is.
- sommige types hebben dus een hogere kans als andere.
