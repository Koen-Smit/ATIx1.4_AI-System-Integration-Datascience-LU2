```python
import seaborn as sns
import matplotlib.pyplot as plt
penguins = sns.load_dataset("penguins")
penguins = penguins.dropna()

# choose: Species, island or sex.
species_means = penguins.groupby('species').mean(numeric_only=True)
print(species_means)
```

#### Bill Length vs Species

```python
sns.barplot(data=penguins, x='species', y='bill_length_mm', errorbar='sd')
plt.title('Mean Bill Length by Species')
plt.ylabel('Bill Length (mm)')
plt.show()
```

##### Conclusion:
Er is niet echt een groot verschil tussen snavel lengte tussen de verschillende species, "Adelie" is wat kleiner vergeleken met de rest. Confidence interval vrij groot en niet opvallend verschillend tussen de species.


---

#### Bill Depth vs Species

```python
sns.barplot(data=penguins, x='species', y='bill_depth_mm', errorbar='sd')
plt.title('Mean Bill Depth by Species')
plt.ylabel('Bill Depth (mm)')
plt.show()
```

##### Conclusion:
"Adelie" en "Chinstrap species" zijn bijna exact gelijk in snavel diepte en de confidence interval, "Gentoo" is wel een stuk lager dan de rest.

#### Flipper Length vs Species

```python
sns.barplot(data=penguins, x='species', y='flipper_length_mm', errorbar='sd')
plt.title('Mean Flipper Length by Species')
plt.ylabel('Flipper Length (mm)')
plt.show()
```

##### Conclusion:
